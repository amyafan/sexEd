## Actual analyses

First, recode all the questions so that 0 is no (or never) and 1 is yes (or at least once)

```python
# q19
survey = survey.replace({'q19': 2}, 0)

# q20 
survey = survey.replace({'q20': 1}, 0)
survey = survey.replace({'q20': [2,3,4,5]}, 1)

# q21
survey = survey.replace({'q21': [1,2]}, 0)
survey = survey.replace({'q21': [3,4,5,6]}, 1)

# q22
survey = survey.replace({'q22': [1,2]}, 0)
survey = survey.replace({'q22': [3,4,5,6]}, 1)
```

```python
print(survey.q19.mean())
print(survey.q20.mean())
print(survey.q21.mean())
print(survey.q22.mean())
```

```python
# unweighted averages by state

survey.groupby("sitecode").mean()
```

```python
s_q19 = survey[~np.isnan(survey.q19)]
s_q20 = survey[~np.isnan(survey.q20)]
s_q21 = survey[~np.isnan(survey.q21)]
s_q22 = survey[~np.isnan(survey.q22)]
```

```python
q19_w  = s_q19.groupby("sitecode").apply(lambda x: np.average(x.q19, weights = x.weight))
q20_w  = s_q20.groupby("sitecode").apply(lambda x: np.average(x.q20, weights = x.weight))
q21_w  = s_q21.groupby("sitecode").apply(lambda x: np.average(x.q21, weights = x.weight))
q22_w  = s_q22.groupby("sitecode").apply(lambda x: np.average(x.q22, weights = x.weight))
```

```python
states = {
    'AK':'Alaska',
    'AL':'Alabama',
    'AR':'Arkansas',
    'AZB':'Arizona',
    'CA':'California',
    'CO':'Colorado',
    'CT':'Connecticut',
    'DC':'District of Columbia',
    'DE':'Delaware',
    'FL':'Florida',
    'GA':'Georgia',
    'HI':'Hawaii',
    'IA':'Iowa',
    'ID':'Idaho',
    'IL':'Illinois',
    'IN':'Indiana',
    'KS':'Kansas',
    'KY':'Kentucky',
    'LA':'Louisiana',
    'MA':'Massachusetts',
    'MD':'Maryland',
    'ME':'Maine',
    'MI':'Michigan',
    'MN':'Minnesota',
    'MO':'Missouri',
    'MS':'Mississippi',
    'MT':'Montana',
    'NC':'North Carolina',
    'ND':'North Dakota',
    'NE':'Nebraska',
    'NH':'New Hampshire',
    'NJ':'New Jersey',
    'NM':'New Mexico',
    'NV':'Nevada',
    'NY':'New York',
    'OH':'Ohio',
    'OK':'Oklahoma',
    'OR':'Oregon',
    'PA':'Pennsylvania',
    'RI':'Rhode Island',
    'SC':'South Carolina',
    'SD':'South Dakota',
    'TN':'Tennessee',
    'TX':'Texas',
    'UT':'Utah',
    'VA':'Virginia',
    'VT':'Vermont',
    'WA':'Washington',
    'WI':'Wisconsin',
    'WV':'West Virginia',
    'WY':'Wyoming'
}
```

```python
combined = pd.concat([q19_w, q20_w, q21_w, q22_w], axis = 1)
combined.columns = ['q19', 'q20', 'q21', 'q22']
combined.reset_index(inplace = True)
combined['sitecode'] = combined['sitecode'].replace(states)
combined.head()
```

```python
yrbs_by_state_str = dataloc + "yrbs_by_state.csv"

combined.to_csv(yrbs_by_state_str)
```

```python
# merge in public health data 

policy = pd.read_csv(manualloc + "Sex_Ed_Policy.csv")
```

```python
merged = pd.merge(policy, combined, left_on='State', right_on='sitecode')
```
